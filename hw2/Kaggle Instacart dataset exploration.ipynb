```python
import pandas
import matplotlib
import seaborn
%matplotlib inline
```

```python
with open('/Users/drew/Google Drive/uw-datasci/data/kaggle/instacart/products.csv') as f:
    products = pandas.read_csv(f)
products[:3]
```

```python
aisle_counts_df = products.groupby('aisle_id').count().reset_index(drop=False)
aisle_counts = aisle_counts_df.product_id
aisle_counts[:5]
```

```python
aisle_counts.plot(kind='bar', width=1, figsize=(16, 4), edgecolor='slateblue', grid=False)
```

```python
aisle_counts.sort_values(ascending=False).plot(kind='bar', width=1, figsize=(16, 4), edgecolor='slateblue', grid=False)
```

```python
# we can look at how the number of products per aisle is distributed across the marketplace
count_distr = aisle_counts.sort_values(ascending=False).reset_index().product_id
count_distr.plot(kind='line', figsize=(16, 4), grid=False)
print 'products per aisle: min/median/mean/max => %d / %d / %.0f / %d' % (
    count_distr.min(), 
    count_distr.median(), 
    count_distr.mean(), 
    count_distr.max())
```
